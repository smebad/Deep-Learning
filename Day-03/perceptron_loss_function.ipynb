```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating a dataset with 2 classes
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, hypercube=False, random_state=41, class_sep=15)
```

```python
# plotting the dataset
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', s=100, alpha=0.6)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show() 
```

```python
# defining functions for the perceptron
def perceptron(X,y):

  w1 = w2 = b = 1
  learning_rate = 0.1

  for j in range(1000):
    for i in range(X.shape[0]):

      z = w1*X[i][0] + w2*X[i][1] + b # calculate the output

      if z*y[i] <= 0:
        w1 = w1 + learning_rate*y[i]*X[i][0] # update the weights
        w2 = w2 + learning_rate*y[i]*X[i][1]
        b = b + learning_rate*y[i]

  return w1, w2, b
```

```python
w1,w2,b = perceptron(X,y) # training the perceptron
```

```python
print(w1)
print(w2)
print(b)
```

```python
m = -(w1/w2) # slope of the decision boundary
c = -(b/w2) # y-intercept of the decision boundary

print(m)
print(c)
```

```python
x_input = np.linspace(-3,3,100) # generating x values for the decision boundary
y_output = m*x_input + c # calculating y values for the decision boundary

# plotting the decision boundary
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_output, color='red', linewidth=2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=100)
plt.ylim(-3, 2)
```
